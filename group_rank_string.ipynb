```python
import pandas as pd

from sqlalchemy import create_engine

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)
pd.set_option('display.max_colwidth', None)



engine = create_engine('mysql+mysqlconnector://root:harsh009%40@127.0.0.1:3306/titanic')

df_marvel = pd.read_sql("SELECT * FROM marvel", con=engine)
df_minfo = pd.read_sql("SELECT * FROM marvel_info", con=engine)

df_dc = pd.read_sql("SELECT * FROM dc", con=engine)
df_dinfo = pd.read_sql("SELECT * FROM dc_info", con=engine)

marvel_full = pd.merge(df_marvel, df_minfo, on="movieid", how = "left")
dc_full = pd.merge(df_dc, df_dinfo, on="movieid", how = "left")

combined = pd.concat([marvel_full, dc_full])

filtered = combined[combined["cinematographer"].str.contains("J",case = False,na = False)]


display(filtered)

```

After merging marvel + marvel_info on movie_id, count how many movies are "BlockBuster" for each director.

```python
import pandas as pd

from sqlalchemy import create_engine

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)
pd.set_option('display.max_colwidth', None)



engine = create_engine('mysql+mysqlconnector://root:harsh009%40@127.0.0.1:3306/titanic')

df_marvel = pd.read_sql("SELECT * FROM marvel", con=engine)
df_minfo = pd.read_sql("SELECT * FROM marvel_info", con=engine)

df_dc = pd.read_sql("SELECT * FROM dc", con=engine)
df_dinfo = pd.read_sql("SELECT * FROM dc_info", con=engine)

marvel_full = pd.merge(df_marvel, df_minfo, on="movieid", how = "left")
dc_full = pd.merge(df_dc, df_dinfo, on="movieid", how = "left")


BBcount = marvel_full[marvel_full["verdict"] == "Blockbuster"].groupby("director")["verdict"].count().sort_values(ascending = False).reset_index()



display(BBcount)
```
